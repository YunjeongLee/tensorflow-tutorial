# Main Code
## Declare parameters

```python
import tensorflow as tf
x = tf.placeholder(tf.float32, [None, 784])
W = tf.Variable(tf.zeros([784, 10]))
b = tf.Variable(tf.zeros([10]))
y = tf.matmul(x,W) + b
```

- `reduce_sum` : adds the elements in the second dimension of y
- `reduce_mean` : computes the mean over all the examples in the batch
- `GradientDescentOptimizer`

```python
y_ = tf.placeholder(tf.float32, [None, 10])
```

- `softmax_cross_entropy_with_logits(logits, labels, dim=-1, name=None)` : computes softmax cross entropy between logits and labels.
- Many classification operations : https://www.tensorflow.org/api_docs/python/nn/classification

```python
learning_rate = 0.5
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(y, y_))
train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cross_entropy)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

## Train
This is called stochastic gradient descent method.

```python
for _ in range(1000):
    batch_xs, batch_ys = mnist.train.next_batch(100)
    sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
```

```python
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
```

```python
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
print(sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
```
